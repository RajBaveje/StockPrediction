```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
#setup
#code
ticker = yf.Ticker('HUM') #import data
aapl_df = ticker.history(period="5y") #get data from 5 year period in dataframe
aapl_df.drop(['High','Low','Volume','Dividends','Stock Splits'], axis=1, inplace=True) #only have two columns, not seven
data = np.empty(shape = (1259), dtype = float)
data_o = aapl_df[['Open']].to_numpy()
data_c = aapl_df[['Close']].to_numpy()
```

```python
X = np.zeros((1259,3))
Y = [0]*1259

for i in range(len(data)-10):
  X[i] = [data_o[i][0], data_c[i][0], data_o[i+1][0]]
  Y[i] = data_o[i+2][0]


X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33)

regr = linear_model.LinearRegression()

regr = regr.fit(X_train, Y_train)
Y_pred = regr.predict(X_test)

print("Mean squared error: %.2f" % mean_squared_error(Y_test, Y_pred))


```

```python
neural_regr = MLPRegressor(random_state=1, max_iter=500).fit(X_train, Y_train)
neural_regr = neural_regr.fit(X_train, Y_train)
neural_Y_pred = neural_regr.predict(X_test)

print("Mean squared error: %.2f" % mean_squared_error(Y_test, neural_Y_pred))
```

```python
random_forest_regr = RandomForestRegressor(random_state=1)
random_forest_regr = random_forest_regr.fit(X_train, Y_train)
Y_pred_rf = random_forest_regr.predict(X_test)

print("Mean squared error: %.2f" % mean_squared_error(Y_test, Y_pred_rf))
```
